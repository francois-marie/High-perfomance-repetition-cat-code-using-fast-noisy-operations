```python
import matplotlib.pyplot as plt
from qsim.utils.quantum_guidelines import global_setup, plot_setup
import pandas as pd
import seaborn as sns
from qsim.logical_gate.process_results import (
    AsymParityMeasurement,
    pz_update_res,
    loading_results,
)
```

```python
method = AsymParityMeasurement()
method.k2a_l = [1, 5, 10, 15, 20, 25]
method.nbar_l = [4, 6, 8, 10, 12, 14, 16]
method.max_p_L = 5e-2
method.date_stamp = '20220914'
```

## compute method

```python
method.dfz = pd.DataFrame(
    pz_update_res(
        loading_results(
            date_stamp=method.date_stamp,
            GATE=method.GATE_pZ,
            distance_l=None,
            suffix='',
            prefix='_distance',
        )
        + loading_results(
            date_stamp=method.date_stamp,
            GATE=method.GATE_pZ,
            distance_l=[i for i in range(0, 1427)],
            suffix='',
        )
    )
)
```

```python
method.dffz = method.compute_dffz()
```

```python
method = AsymParityMeasurement()
method.load_df()
```
